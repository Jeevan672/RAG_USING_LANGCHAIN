```python
from langchain.text_splitter import MarkdownHeaderTextSplitter
```

```python
mark_doc="""# Title\n\n \
## Chapter 1\n\n \
Hi this is Jeeevan\n\n Hi this is Reddy\n\n \
### Section \n\n \
Hi this is Gade \n\n
## chapter 2\n\n \
Hi this is Jeevan"""
```

```python
headers_to_split_on=[
    ("#","Header 1"),
    ("##","Header 2"),
    ("###","Header 3"),
]
```

```python
markdown_splitter=MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on)
```

```python
md_header_splits=markdown_splitter.split_text(mark_doc)
```

```python
md_header_splits[0]
```
